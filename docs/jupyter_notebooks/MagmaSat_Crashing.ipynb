## Solving MagmaSat Crashes
- This isn't an ideal fix, but allows you to see how to troubleshoot

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import VESIcal as v
```

```python
myfile=v.BatchFile('Example_Crash.xlsx', sheet_name="ExampleCrashing",  input_type='wtpercent')
```

```python
mydata=myfile.get_data()
mydata
```

```python
# This cell crashes as it prints [===============     ] 75%  Working on sample test3        
#myfile.calculate_saturation_pressure(temperature=1250)
```

```python
# You can see here it is indeed test3 that is breaking it
mydata1_ind=mydata.iloc[2].to_frame().T
myfile1_ind=v.BatchFile(filename=None, label=mydata1_ind.index, dataframe=mydata1_ind)
myfile1_ind.calculate_saturation_pressure(temperature=1240)
```

```python
new= mydata.copy()
filt=new[new.index!='test3'] # Exclude this one. 
myfile2=v.BatchFile(filename=None, label=filt.index, dataframe=filt)
out=myfile2.calculate_saturation_pressure(temperature=1250)
out
```
