```python
import VESIcal as v
import matplotlib.pyplot as plt
import numpy as np

myfile = v.ExcelFile('Moore_et_al_1998.xlsx', input_type='wtpercent')
#myfile.data
```

```python
MooreFluid = myfile.calculate_equilibrium_fluid_comp(pressure="Pressure_bars", temperature="Temperature_C", model='MooreWater')
MooreFluid
```

```python
x = np.linspace(0,1,100)
y = x
ax = plt.subplot()
ax.plot(MooreFluid["XH2Ofl_reported"], MooreFluid["XH2O_fl_VESIcal"], 'o', color='black')
ax.plot(x, y, '-')
ax.set_xlabel("XH2Ofl_reported")
ax.set_ylabel("XH2Ofl_calculated")
ax.set_title("MooreWater().calculate_equilibrium_fluid_comp()")
plt.show()
```

```python
MooreVolatiles = myfile.calculate_dissolved_volatiles(pressure="Pressure_bars", temperature="Temperature_C", X_fluid="XH2Ofl_reported", model='MooreWater')

x = np.linspace(0,10,100)
y = x
ax = plt.subplot()
ax.plot(MooreVolatiles["H2O"], MooreVolatiles["H2O_liq_VESIcal"], 'o', color='black')
ax.plot(x, y, '-')
ax.set_xlabel("H2O wt% reported")
ax.set_ylabel("H2O wt% calculated")
ax.set_title("MooreWater().calcualte_dissolved_volatiles()")
plt.show()
```

```python
MoorePressures = myfile.calculate_saturation_pressure(temperature="Temperature_C", X_fluid="XH2Ofl_reported", model='MooreWater')
MoorePressures
x = np.linspace(0,6000,100)
y = x
ax = plt.subplot()
ax.plot(MoorePressures["Pressure_bars"], MoorePressures["SaturationP_bars_VESIcal"], 'o', color='black')
ax.plot(x, y, '-')
ax.set_xlabel("Experimental Pressure reported")
ax.set_ylabel("Saturation Pressure calculated")
ax.set_title("MooreWater().calcualte_saturation_pressure()")
plt.show()
```
