```python
satPs = pd.read_excel('satPs.xlsx', sheet_name='Calc0')
satPs = satPs.fillna(' ')
satPs = satPs.rename(columns={'Unnamed: 0': 'Label'})
satPs = satPs.drop(['SiO2', 'TiO2', 'Al2O3', 'Fe2O3', 'Cr2O3', 'FeO', 'MnO', 'MgO', 'NiO', 'CoO', 'CaO', 'Na2O', 'K2O', 'P2O5', 'H2O', 'CO2', 
                    'CITATION', 'ROCK TYPE', 'Press', 'Temp'], axis=1)

# make column of '--'
dash_col = []
for r in range(len(satPs)):
    dash_col.append('--')

# add that column to the dataframe
columns = ['User Input Data'] + list(satPs.columns.values)
satPs['User Input Data'] = dash_col
satPs = satPs[columns]

# add spaces to some col names
satPs = satPs.rename(columns={'SaturationP_bars_VESIcal': 'SaturationP_bars _VESIcal',
                              'Temperature_C_VESIcal': 'Temperature_C _VESIcal',
                              'XH2O_fl_VESIcal': 'XH2O_fl _VESIcal',
                              'XCO2_fl_VESIcal': 'XCO2_fl _VESIcal',
                              'FluidMass_grams_VESIcal': 'FluidMass _grams_VESIcal',
                              'FluidSystem_wt_VESIcal': 'FluidSystem _wt_VESIcal'})

satPs.set_index('Label')
```
